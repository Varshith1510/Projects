```python
#importing
import pandas as pd
df=pd.read_csv(r"C:\Users\Micro\Desktop\SEM 3\Mini project\IPL Ball by Ball.csv")
df.info()
```

```python
#max runs over seasons by teams
df1=pd.pivot_table(df,columns="batting_team",values="run",index="season",aggfunc=lambda x:x.sum())
for i in df1:
    print(i,"-",int(df1[i].max()))
```

```python
#max runs over seasons by players
df2=pd.pivot_table(df,columns="striker",values="runs_off_bat",index="season",aggfunc=lambda x:x.sum())
for i in df2:
    print(i,"-",int(df2[i].max()))
```

```python
#avg runs in powerplay considering seasons
df3=df[(df["over"]>=0) & (df["over"]<=5)]
df4=pd.pivot_table(df3,index="season",columns="batting_team",values="runs_off_bat",aggfunc=lambda x:x.sum())
for i in df4:
    print(i,"-",int(df4[i].mean()))
```

```python
#Mamimum no of 50 over matches in all seasons
df5=pd.pivot_table(df,columns="match_id",index="striker",values="runs_off_bat",aggfunc=lambda x:x.sum())
df5["century"]=0
l1,l2=[],[]
l1=df5.index
l2=df5.columns
for i in l1:
    for j in l2:
        if(df5.loc[i,j]>=100):
            df5.loc[i,"century"]+=1
df6=df5[df5["century"]==df5["century"].max()]
df6["century"]
```
